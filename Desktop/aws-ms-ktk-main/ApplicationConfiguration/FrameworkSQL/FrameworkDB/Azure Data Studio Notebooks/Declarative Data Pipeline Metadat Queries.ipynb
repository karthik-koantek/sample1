**Get List of Projects**

```sql
SELECT * FROM dbo.Project;
```

**Get List of Project Runs**

```sql
EXEC dbo.GetProjectRuns;
```

**Get Run Details of One Project Run**

```sql
EXEC dbo.GetProjectRunDetail @ProjectLogKey = 28;
```

**Get just the details of any failed steps**

```sql
EXEC dbo.GetProjectRunDetail @ProjectLogKey = 28, @LogStatus = 'Failed';
```

**Get Declarative Pipeline Configuration for a Project**

```sql
EXEC dbo.GetProjectConfiguration @ProjectName = 'Data Catalog';
```
